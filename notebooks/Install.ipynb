# Installing *heptabot* on raw system

This notebook was tested on [vast.ai](https://vast.ai/console/create/) instances using `tensorflow/tensorflow:2.3.0-gpu-jupyter` image. *heptabot* is supposed to be run on Linux x64 arhitecture.

First we install packages to interact with web repositories where we store our code and models: `git` and `wget`. Here and below, `yes` channels the "yes" message to every command call to agree with all prompts.

```python
!yes | apt install git wget
```

Now we get the *heptabot* code and enter the directory.

```python
!git clone https://github.com/lcl-hse/heptabot
%cd heptabot
```

To facilitate the installation of all necessary packages this script relies on `conda`. So now we check if `conda` is installed, and, as it isn't, we let the code install it for us.

```python
!yes | bash setup.sh
```

Now that `conda` is installed, we need to reconnect to terminal so it acknowledges the package.

### Hit `Kernel` -> `Bash` after this step

In case you don't have it installed, run `pip install bash_kernel && python -m bash_kernel.install` from your terminal first.

We enter the heptabot folder once again and then run a series of commands so that we could measure the time our installation process will take.

```python
cd heptabot
touch run_setup.sh
echo "yes | bash setup.sh" > run_setup.sh
chmod +x run_setup.sh
```

Now, finally, we can execute the rest of the script that will create the `heptabot` virtual environment and install *heptabot* there. Keep in mind that the time it will take to install (obviously) depends on the network speed and computational capabilites of your system and that the first `heptabot` run will take a longer time (as it will download the model needed for `sentence_transformers` etc.)

```python
time ./run_setup.sh
```

After this step, if you go to Jupyter starting page and hit F5, `heptabot` should appear in the list of your kernels.
